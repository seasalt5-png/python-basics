# Data Structures in Python
---

Data structures are specialized formats used to store and organize data efficiently.  In Python, several built-in data structures allow programmers to manage collections of values in different ways. Understanding data structures is essential for data manipulation, algorithm design, and data analysis tasks.

Python provides several built-in data structures, including:

- Lists
- Tuples
- Sets
- Dictionaries

Each structure has different properties and is suited for different types of problems.

## 1. Lists

A list is an ordered collection of items. Lists are one of the most commonly used data structures in Python.

Characteristics of lists:

- Ordered
- Mutable (can be modified)
- Allow duplicate values
- Can store multiple data types

**Lists are created using square brackets [  ]**.

*note:  
Lists use indexing to access elements. Indexing starts at 0.*

```python
# example

fruits = ["apple", "banana", "orange"]
print(fruits)
```

Common list methods include:

- list[i] → access element at index i
- list[start:stop] → slice from start to stop-1
- list[-1] → access last element
- append(x) → add x to the end  
- insert(i, x) → insert x at index i 
- pop(i) → remove and return item at i  
- sort() → sort list in ascending order  
- len() → get number of elements  
- copy() → create a shallow copy  
- in → check if item exists (True/False)  
- (+)→ concatenate (join) two lists
- (* )→ repeat list n times


```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

print(fruits[1])          # Access element at index 1
print(fruits[1:4])        # Slice from index 1 to 3
print(fruits[-1])         # Access last element 

fruits.append("fig")      # Add "fig" to the end
fruits.insert(2, "grape") # Insert "grape" at index 2
popped_item = fruits.pop(0) # Remove and return item at index 0
fruits.sort()             # Sort list in ascending order
length = len(fruits)      # Get number of elements
new_list = fruits.copy()  # Create a shallow copy
combined = fruits + [1, 2] # Concatenate (join) two lists
repeated = ["A"] * 3      # Repeat list 3 times: ["A", "A", "A"]

print(f"Final list: {fruits}")
```

## 2. Tuples

A tuple is similar to a list but immutable, meaning its values cannot be modified after creation.

Characteristics:

- Ordered
- Immutable
- Allows duplicate values

**Tuples are created using parentheses ( )**.

```python
# example

coordinates = (10, 20)
print(coordinates)
```

---

## 3. Sets

A set is an unordered collection of unique elements.

Characteristics:

- Unordered
- No duplicate values
- Useful for removing duplicates

**Sets are created using curly braces { }**

```python
# example

numbers = {1, 2, 3, 3, 4, 3, 1}
print(numbers)
```

---

## 4. Dictionaries

A dictionary stores data in key-value pair.

Characteristics:

- Unordered
- Mutable
- Keys must be unique

**Dictionaries are created using curly braces { }.**

```python
# example

student = {
    "name": "Alice",
    "age": 21,
    "major": "Data Science"
}
print(student, "\n")

# acessing dictionary value
print(student["name"])
print(student["age"], "\n")

# adding or updating values
student["age"] = 22
student["gpa"] = 3.8
print(student, "\n")

# looping
student = {
    "name": "Alice",
    "age": 21,
    "major": "Data Science"
}

for key, value in student.items():
    print(key, ":", value)
```

## Mini Challenge

Create a program that counts how many times each word appears in a sentence.

Example sentence:

"data science is fun and data science is powerful"

```python
sentence = "data science is fun and data science is powerful"
words = sentence.split()
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```
