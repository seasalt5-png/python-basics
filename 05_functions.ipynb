# Functions in Python

Functions are reusable blocks of code designed to perform a specific task. Instead of repeating the same code multiple times, functions allow programmers to organize logic into manageable and reusable components.

Functions improve code readability, reduce repetition, and make programs easier to maintain. They are widely used in data science workflows, software development, and automation tasks.

In this notebook, we will learn how to create and use functions in Python, including passing arguments, returning values, and writing reusable functions for practical problems.

## 1. What is a Function?

A function is a reusable block of code that performs a specific task. Instead of rewriting the same instructions multiple times, we can define a function once and call it whenever needed.

Functions help:

- organize programs into smaller components
- reduce repetitive code
- improve readability and maintainability

## 2. Defining a Function

Functions in Python are defined using the **def**keyword.

General syntax:

def function_name():  
    # code block

After defining a function, it can be executed by calling the function.

```python
# example

def greet():
    print("Hello, welcome to Python!")

greet()
```

## 3. Functions with Parameters

Functions can accept parameters, which are inputs passed into the function.

Parameters allow the function to work with different values.

```python
# example

def greet_user(name):
    print("Hello", name)

greet_user("Alice")
greet_user("Bob")
```

## 4. Returning Values

Functions often perform calculations and return results. This is done using the **return** statement.

The returned value can then be stored in a variable or used in further calculations.

```python
# example

def add_numbers(a, b):
    result = a + b
    return result

sum_value = add_numbers(5, 3)
print(sum_value)
```

## 5. Functions with Multiple Parameters

Functions can accept multiple parameters to perform more complex tasks.

```python
# example

def calculate_area(length, width):
    area = length * width
    return area

print(calculate_area(5, 3))
```

## 6. Default Parameters

Functions can also define **default values** for parameters. If the argument is not provided, the default value will be used.  

```python
# example

def greet(name="Patrick"):    # Default value from this function is "Patrick"
    
    print("Hello", name)

greet("Rose")
greet()
```

Create a function called "temperature_converter".
