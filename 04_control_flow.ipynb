# Control Flow in Python (if, elif, else)

Control flow statements allow a program to make decisions based on conditions. Instead of executing every line of code sequentially, programs can choose different paths depending on whether certain conditions are true or false.

In Python, the most common control flow statements are:

- if
- elif
- else

These statements evaluate logical conditions and determine which block of code should be executed.

Understanding control flow is essential for building intelligent programs that react to different inputs and situations.

## 1. What is Control Flow?

In programming, control flow determines the order in which statements are executed.

Normally, a Python program runs from top to bottom. However, with control flow statements, we can control which parts of the code are executed depending on certain conditions.

*note:   
Python uses indentation to define blocks of code. Proper indentation is required for the program to run correctly.*

## 2. The if  Statement

The if statement executes a block of code only if the specified condition is **True**.

This is commonly used to check values, validate inputs, or make decisions in a program.

```python
# example

age = 20
 
if age >= 18:
    print("You are an adult.")
```

## 3. The else Statement

The else statement defines a block of code that will execute if the condition in the if statement is **False**.

This allows programs to handle alternative outcomes.

```python
# example

age = 16

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

## 4. The elif Statement

The elif statement (short for **else if**) allows us to check multiple conditions sequentially.

If the first condition is False, Python evaluates the next condition.

```python
# example

score = 85

if score >= 90:
    print("Grade: A")
elif score >= 75:
    print("Grade: B")
elif score >= 60:
    print("Grade: C")
else:
    print("Grade: D")
```

## 5. Combining Conditions

Conditions can be combined using logical operators such as:

- and
- or
- not

This allows programs to evaluate more complex situations.

```python
# example

age = 22
has_license = True

if age >= 18 and has_license:
    print("You are allowed to drive.")
else:
    print("You cannot drive.")
```

## 6. Nested Conditions

Control flow statements can also be nested, meaning one if statement can appear inside another.

This allows for more detailed decision-making logic.

```python
# example

age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Access granted.")
    else:
        print("ID required.")
else:
    print("Access denied.")
```
