# **Operators in Python**
---

Operators are fundamental components of programming that allow us to perform operations on variables and values.  
In Python, operators are used to perform tasks such as mathematical calculations, comparisons, and logical decisions.  
Understanding operators is essential because they are widely used in data processing, algorithm development, and decision-making logic within programs.

## 1. What is an Operator?

An operatoris a symbol used to perform an operation on one or more values (called operands).

For example, in the expression:  
5 + 3


- 5 and 3 are operands 
- (+) is the operator 

Python supports several categories of operators, including:

- Arithmetic Operators
- Comparison Operators
- Logical Operators
- Assignment Operators

## 2. Arithmetic Operators

Arithmetic operators are used to perform mathematical calculations.

Common arithmetic operators in Python include:

| Operator | Description | Example |
|--------|--------|--------|
| + | Addition | 5 + 3 |
| - | Subtraction | 5 - 3 |
| * | Multiplication | 5 * 3 |
| / | Division | 5 / 3 |
| // | Floor Division | 5 // 3 |
| % | Modulus (remainder) | 5 % 3 |
| ** | Exponentiation | 5 ** 3 |

```python
a = 5
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

## 3. Comparison Operators

Comparison operators are used to compare two values. The result of a comparison is always a Boolean value **(True or False).**

| Operator | Meaning |
|--------|--------|
| == | Equal to |
| != | Not equal to |
| > | Greater than |
| < | Less than |
| >= | Greater than or equal to |
| <= | Less than or equal to |

```python
x = 10.5
y = 5

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

## 4. Logical Operators

Logical operators are used to combine multiple conditions. They are commonly used in decision-making statements such as if conditions.

Python provides three logical operators:

| Operator | Description |
|--------|--------|
| and | Returns True if both conditions are true |
| or | Returns True if at least one condition is true |
| not | Reverses the Boolean result |

```python
age = 20
has_id = True

print(age > 18 and has_id) 
print(age > 21 or has_id)
print(not has_id)
```

## 5. Assignment Operators

Assignment operators are used to assign values to variables.

The most common assignment operator is **=**. However, Python also supports combined assignment operators, which perform an operation and assignment in one step.

| Operator | Example | Meaning |
|--------|--------|--------|
| = | x = 5 | Assign value |
| += | x += 3 | x = x + 3 |
| -= | x -= 3 | x = x - 3 |
| *= | x *= 3 | x = x * 3 |
| /= | x /= 3 | x = x / 3 |

```python
score = 10

score += 5
print(score)

score *= 2
print(score)

score -= 4
print(score)
```
