# **Variables and Data Types in Python**  



Understanding variables and data types is one of the most fundamental concepts in Python programming. Every program stores, modifies, and processes data. In Python, this data is stored in variables, and each variable contains a specific data type that determines how the value can be used.

This notebook introduces the concept of variables, basic data types, and how Python handles them. The examples progress from simple usage to slightly more practical cases.

## **1. Introduction to Variables**  

A variable is a name used to store data in memory. Instead of repeatedly writing the same value, we store it in a variable and reuse it whenever needed.

Python is a dynamically typed language, meaning you do not need to declare the data type explicitly. The interpreter automatically determines the type based on the assigned value.

General syntax:  

variable_name = value

A good variable name should:

- Be descriptive
- Use lowercase letters
- Separate words using underscores

```python
# Creating variables

name = "Alice"
age = 25
height = 1.65

print(name)
print(age)
print(height)
```

## **2. Variable Naming Rules**  

When creating variables in Python, several rules must be followed:

### Valid Rules

- Variable names must start with a letter or underscore
- They may contain etters, numbers, and underscores
- Python is case-sensitive

Example:  
    user_name  
    age  
    total_score

Invalid Examples:  
    2name  
    user-name  
    class  

***Avoid using Python reserved keywords as variable names.***

```python
# Valid variable names

user_name = "John"
total_score = 95
item_count = 10

print(user_name, total_score, item_count)
```

## **3. Basic Data Types**
There are various data types in programming, but these are the fundamental types commonly used:

- Integer (int): Used for whole numbers without a fractional part, such as 10 or -5.

-  Float (float): Used for real numbers that contain a decimal point, such as 3.14 or 2.0.

- String (str): Used for sequences of characters or text, defined using single or double quotes.

- Boolean (bool): Used to represent one of two values: True or False, typically resulting from logical comparisons.

```python
# Different data types

integer_example = 10
float_example = 3.14
string_example = "Hello Python"
boolean_example = True

print(integer_example)
print(float_example)
print(string_example)
print(boolean_example)
```

## 4. Checking the Data Type

Python provides a built-in function called **type()** that allows us to check the data type of a variable.

This is useful when working with data analysis or debugging code.

```python
x = 10
y = 3.5
z = "Data Science"
a = False

print(type(x))
print(type(y))
print(type(z))
print(type(a))
```

## 5. Type Conversion

Sometimes we need to convert one data type into another. This process is called **type casting**.

Common conversions include:

- int() → convert to integer
- float() → convert to float
- str() → convert to string

Type conversion is important when working with user input or numerical calculations.

```python
# Type conversion examples

num_int = 10
num_float = float(num_int)

print(num_int)
print(num_float)
print(type(num_float))
```

## 6. Simple Practical Example

Variables are commonly used in calculations. Below is a simple example that calculates the total price of purchased items.

```python
item_price = 15.5
quantity = 3

total_price = item_price * quantity

print("Total price:", total_price)
```
